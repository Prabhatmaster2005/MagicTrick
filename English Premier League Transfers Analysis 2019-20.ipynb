```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")
```

```python
#Total value of the purchases made by each team
file2 = 'EPL_team.xlsx'
df2 = pd.read_excel(file2)
df2.head()
```

```python
#How much money each club spent 
sns.barplot(x="club_name",
            y="Total value of purchase",
            data=df2,
            palette="rocket",
            order=df2.sort_values('Total value of purchase',ascending = False).club_name)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 30,rotation = 65)
plt.yticks(fontsize = 30)
plt.xlabel("Club",fontsize = 35)
plt.ylabel("Total value of purchase in £ million",fontsize = 35)
plt.title("Total value of purchase based on club bought",fontsize = 50)
```
