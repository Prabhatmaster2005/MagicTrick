```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file  = '2018 FIFA World Cup Squads.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['Team']
av = df['Team'].value_counts()
plt.plot(av)
print(av)
plt.show()
```

```python
b = df['Group']
bv = df['Group'].value_counts()
plt.plot(bv)
plt.show()
```

```python
c = df['Age']
cv = df['Age'].value_counts()
plt.plot(cv,c ='Black' , ls='', marker = 'o', ms=14)
print(cv)
plt.show()
```

```python
d = df['Caps']
dv = df['Caps'].value_counts()
plt.plot(dv,c ='Black' , ls='', marker = 'o', ms=14)
print(dv)
plt.show()
```

```python
e = df['Goals']
ev = df['Goals'].value_counts()
plt.plot(ev,c ='Black' , ls='', marker = 'o', ms=14)
print(ev)
plt.show()
```

```python
f = df['Club']
fv = df['Club'].value_counts()
fd = dict(fv)
print(fd)
```

```python
g = df['Squad Number']
gv = df['Squad Number'].value_counts()
plt.plot(gv,c ='Black' , ls='', marker = 'o', ms=20)
print(gv)
plt.show()
```
