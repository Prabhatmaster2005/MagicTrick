```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file  = '2018 FIFA World Cup Squads.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['Team']
av = df['Team'].value_counts()
plt.plot(av,c ='Red' , ls='', marker = 'o', ms=20)
print(av)
plt.xticks(fontsize = 25,rotation = 70)
plt.yticks(fontsize = 25)
plt.ylabel("Number of players ",fontsize = 25)
plt.xlabel("Team",fontsize = 25)
plt.title("Number of players in each team",fontsize = 50)
plt.show()
```

```python
c = df['Age']
cv = df['Age'].value_counts()
plt.plot(cv,c ='Blue' , ls='', marker = 'o', ms=20)
print(cv)
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25)
plt.ylabel("Number of players ",fontsize = 25)
plt.xlabel("Age",fontsize = 25)
plt.title("Number of players in each age",fontsize = 50)
plt.show()
```
