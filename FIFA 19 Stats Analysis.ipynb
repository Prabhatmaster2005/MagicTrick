```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'FIFA 19 Stats.csv'
df= pd.read_csv(file)
df.head()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['Position']
av = df['Position'].value_counts()
a_dict = list(dict.fromkeys(a))
plt.plot(av,c ='Red' , ls='', marker = 'o', ms=25)
plt.show
plt.xticks(fontsize = 25,rotation = 65)
plt.yticks(fontsize = 25)
plt.xlabel("Position",fontsize = 25)
plt.ylabel("Number of players",fontsize = 25)
plt.title("Number of players sold based on position",fontsize = 40)
av.head(10)
```
