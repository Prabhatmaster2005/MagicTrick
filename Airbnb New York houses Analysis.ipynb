```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'AB_NYC_2019.csv'
df = pd.read_csv(file)
df.head()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['neighbourhood_group']
av = a.value_counts()
a_dict = list(dict.fromkeys(a))
plt.plot(av,c ='Blue' , ls='', marker = 'o', ms=30)
plt.show
plt.xticks(fontsize = 25,rotation = 65)
plt.yticks(fontsize = 25)
plt.xlabel("Neighbourhood group",fontsize = 25)
plt.ylabel("Number of houses",fontsize = 25)
plt.title("Number of houses based on neighbourhood group",fontsize = 40)
print(av)
```
