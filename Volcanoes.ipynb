```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'volcano.csv'
df = pd.read_csv(file)
df.head()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df["Country"]
av = a.value_counts()
av.head()
```

```python
plt.plot(av.head(20),c ='Red' , ls='', marker = 'o', ms=20)
plt.xticks(fontsize = 25, rotation = 65)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.yticks(fontsize = 25)
plt.ylabel("Number of Volcano Eruptions",fontsize = 25)
plt.xlabel("Country",fontsize = 25)
plt.title("Number of Volcano Eruptions per Country",fontsize = 40)
```

```python
b = df["Region"]
bv = b.value_counts()
bv.head()
```

```python
plt.plot(bv.head(20),c ='Blue' , ls='', marker = 's', ms=20)
plt.xticks(fontsize = 25, rotation = 65)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.yticks(fontsize = 25)
plt.ylabel("Number of Volcano Eruptions",fontsize = 25)
plt.xlabel("Region",fontsize = 25)
plt.title("Number of Volcano Eruptions per Region",fontsize = 40)
```
