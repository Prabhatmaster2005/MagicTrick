```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'Pokemon.csv'
df = pd.read_csv(file)
df.head()
```

```python
df['Name']
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['Type 1'].value_counts()
a
plt.plot(a,c ='Black' , ls='', marker = 'o', ms=30)
plt.xticks(fontsize = 35,rotation = 65)
plt.yticks(fontsize = 35)
plt.ylabel("Number of pokemon ",fontsize = 35)
plt.xlabel("Type 1",fontsize = 35)
plt.title("Number of pokemon in type 1",fontsize = 50)
```

```python
b = df['Type 2'].value_counts()
plt.plot(b,c ='Blue' , ls='', marker = 'o', ms=30)
plt.xticks(fontsize = 35,rotation = 65)
plt.yticks(fontsize = 35)
plt.ylabel("Number of pokemon ",fontsize = 35)
plt.xlabel("Type 2",fontsize = 35)
plt.title("Number of pokemon in type 2",fontsize = 50)
```

```python
c = df['Generation'].value_counts()
c = sorted(c)
plt.plot(c,c ='Red' , ls='', marker = 'o', ms=40)
plt.xticks(fontsize = 35)
plt.yticks(fontsize = 35)
plt.ylabel("Number of pokemon ",fontsize = 35)
plt.xlabel("Generation",fontsize = 35)
plt.title("Number of pokemon each generation",fontsize = 50)
```
