```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'Most expensive transfers.csv'
```

```python
df = pd.read_csv(file)
df.head(5)
```

```python
z = df['Player'].count()
z
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,15
```

```python
a = df['Fee(€ mln)']
b = df['Year']
c = df['Position']
d = df['Fee(£ mln)']
e = df['From']
f = df['Year of birth']
g = df['To']
```

```python
a = df['Fee(€ mln)']
av = df['Fee(€ mln)'].value_counts()
a_dict = list(dict.fromkeys(a))
plt.plot(av,c ='Black' , ls='', marker = 'o', ms=14)
plt.xticks(fontsize = 20,rotation=70)
plt.yticks(fontsize = 20)
plt.xlabel("Fee(in € mln)",fontsize = 25)
plt.ylabel("Player",fontsize = 25)
plt.title("Players sold based on Fee",fontsize = 40)
plt.show()
```

```python
dv = d.value_counts()
d_dict = list(dict.fromkeys(d))
plt.plot(dv,c ='Black' , ls='', marker = 'o', ms=14)
plt.xticks(fontsize = 20,rotation=70)
plt.yticks(fontsize = 20)
plt.xlabel("Fee(Fee(£ mln))",fontsize = 25)
plt.ylabel("Player",fontsize = 25)
plt.title("Players sold based on Fee",fontsize = 40)
plt.show()
```

```python
b = df['Year']
bv = df['Year'].value_counts()
b_dict = list(dict.fromkeys(b))
plt.plot(bv,c ='Blue' , ls='', marker = 'o', ms=30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel("Year",fontsize = 25)
plt.ylabel("Number of players ",fontsize = 25)
plt.title("Number of players based on year",fontsize = 40)
```
