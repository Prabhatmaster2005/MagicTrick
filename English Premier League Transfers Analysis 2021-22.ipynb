# Analysis of the players bought by the Premier League Teams

```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")
```

```python
#Importing the file
file = "https://raw.githubusercontent.com/ewenme/transfers/master/data/2021/english_premier_league.csv"
```

```python
#Reading the file in pandas
df = pd.read_csv(file)
#Showing the first 5 rows
df.head()
```

```python
tidf = df[df['transfer_movement']=='in']
todf = df[df['transfer_movement']=='out']
```

```python
#Finding the total amount of money spent in EPL Transfers 2019/20
z = df['fee_cleaned'].sum()
print("A total of £{:.2f} million was spent in the EPL Transfers 2019/20 seaon".format(z))
```

## Number of players

```python
#Finding the Number of Players sold Based on Position
a = tidf['position']
av = tidf['position'].value_counts()
a_dict = list(dict.fromkeys(a))
sns.countplot(x = "position",
              data = tidf,
              palette="autumn",
              order = tidf['position'].value_counts().index)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 30,rotation = 60)
plt.yticks(fontsize = 30)
plt.xlabel("Position",fontsize = 35)
plt.ylabel("Number of players",fontsize = 35)
plt.title("Number of players bought based on position",fontsize = 50)
print(av)
```

```python
#Finding the Number of Players Bought Based on Fee
b = tidf['fee_cleaned']
bv = tidf['fee_cleaned'].value_counts()
b_dict = list(dict.fromkeys(b))
plt.plot(bv,c ='Magenta' , ls='', marker = 'o', ms=15)
print(bv.head())
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.xlabel("Fee in £ (in million)",fontsize = 35)
plt.ylabel("Number of players",fontsize = 35)
plt.title("Number of players bought based on fee",fontsize = 50)
plt.show()
```
